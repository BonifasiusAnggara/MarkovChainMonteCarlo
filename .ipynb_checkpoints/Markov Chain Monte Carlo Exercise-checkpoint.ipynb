## Fungsi Logistik untuk mewakili transisi

Kita perlu memutuskan suatu fungsi untuk mewakili transisi dari keadaan terbangun menuju tidur. Ada sejumlah model yang dapat diterima, dan disini kita akan menganggap transisi ini bisa dimodelkan sebagai fungsi logistik. Fungsi Logistik (juga disebut sigmoid) adalah fungsi non linier yang dibatasi 0 dan 1. Untuk ${t \to -\infty}, {p(s|t) \to 0}$ dan, as ${t \to +\infty}, {p(s|t) \to 1}$. Ekspresi untuk Distribusi Probabilitas Logistik untuk tidur sebagai fungsi waktu adalah:

$$p(s|t) = \frac{1}{ 1 + e^{\;\beta t} }$$

Parameter $\beta$ tidak diketahui dan diperkirakan menggunakan pengambilan sampel Markov Chain Monte Carlo. Sampel MCMC dari sebelumnya untuk tiap parameter, mencoba untuk memaksimalkan probabilitas dari parameter yang diberikan data.

Beberapa fungsi Logistik dari berbagai parameter $\beta$ ditunjukan dibawah ini:

```python
figsize(16,6)

# Logistic Function with only Beta
def logistic(x, beta):
    return 1./(1. + np.exp(beta*x))

# Plot examples with different beta
x = np.linspace(-5, 5, 1000)
for beta in [-5, -1, 0.5, 1, 5]:
    plt.plot(x, logistic(x, beta), label = r"$\beta$ = %.1f" % beta)
    
plt.legend();
plt.title(r'Logistic Function with Different $\beta$ values');
```

Ada satu masalah dengan fungsi Logistik dasar seperti yang ditunjukkan diatas: transisi dipusatkan di 0. Bagaimanapun, dalam data tidurnya, transisi adalah sekitar pukul 10:00 malam untuk tidur dan 06:00 pagi untuk bangun. Kita mengatasinya dengan menambahkan offset, disebut bias, untuk menyesuaikan lokasi fungsi logistik. Fungsi Logistik sekarang adalah:

$$p(t) = \frac{1}{1 + e^{\;\beta t + \alpha}}$$

Ini memperkenalkan parameter lain yang tidak diketahui, yang juga akan  kita temukan dari Markov Chain Monte Carlo.

Fungsi Logistik dengan berbagai parameter $\alpha$ dan $\beta\$ ditunjukkan dibawah ini:

```python
figsize(20, 8)

# Logistic function with both beta and alpha
def logistic(x, beta, alpha=0):
    return 1./(1. + np.exp(np.dot(beta, x) + alpha))

x = np.linspace(-4, 6, 1000)

plt.plot(x, logistic(x, beta=1), label=r"$\beta = 1, \alpha = 0$", ls="--", lw=2)
plt.plot(x, logistic(x, beta=-1), label=r"$\beta = -1, \alpha = 0$", ls="--", lw=2)
plt.plot(x, logistic(x, -1, 1), label=r"$\beta = -1, \alpha = -1$", color="darkblue")
plt.plot(x, logistic(x, -1, -1), label=r"$\beta = -1, \alpha = -1$", color="skyblue")
plt.plot(x, logistic(x, -2, 5), label=r"$\beta = -2, \alpha = -2$", color="orangered")
plt.plot(x, logistic(x, -2, -2), label=r"$\beta = -2, \alpha = -5$", color="darkred")
plt.legend(); plt.ylabel('Probability'); plt.xlabel('t')
plt.title(r'Logistic Function with varying $\beta$ and $\alpha$');
```

$\beta$ menggeser arah dan kecuraman kurva. sementara $\alpha\$ mengubah lokasi. Kita akan menggunakan MCMC untuk menemukan nilai yang paling mungkin dari parameter ini dibawah data.

## Distribusi sebelumnya dari $\beta\$ dan $\alpha\$

Kita tidak memiliki bukti untuk menyarankan distribusi sebelumnya untuk parameter model $\beta\$ dan $\alpha\$. Oleh karena itu, kita bisa memodelkannya seolah-olah mereka berasal dari distribusi normal. Distribusi Normal, atau Gaussian, didefinisikan oleh mean, $\mu\$, dan presisi, $\tau\$. Presisi adalah kebalikan dari standar deviasi, $\sigma\$. Mean mendefinisikan lokasi distribusi dan presisi menunjukkan penyebarannya. Nilai $\tau\$ yang lebih besar menunjukkan data kurang tersebar (lebih presisi) dan karenanya variasinya lebih kecil. Mean bisa positif atau negatif, tapi presisi akan selalu positif. Distribusi normal sebagaimana didefinisikan disini direpresentasikan sebagai:

$$ f(x | \mu, \tau) = \sqrt{\frac{\tau}{2\pi}} \exp\left( -\frac{\tau}{2} (x - \mu)^2 \right) $$

Fungsi Kepadatan Probabilitas untuk tiga distribusi normal ditunjukkan dibawah ini:

```python
figsize(20, 8)

# Setup the plotting parameters
nor = stats.norm
x = np.linspace(-10, 10, 1000)
mu = (-5, 0, 4)
tau = (0.5, 1, 2.5)
colors = ("forestgreen", "navy", "darkred")

# Plot 3 pdfs for different normal distributions
params = zip(mu, tau, colors)
for param in params:
    y = nor.pdf(x, loc = param[0], scale = 1 / param[1])
    plt.plot(x, y, label = "$\mu = %d,\;\\tau = %.1f$" % (param[0], param[1]),
            color = param[2])
    plt.fill_between(x, y, color = param[2], alpha=0.3)
    
plt.legend(prop={'size':18});
plt.xlabel('$x$');
plt.ylabel("Probability Density", size=18);
plt.title("Normal Distributions", size=20);
```

## Parameter Search Space

```python
%matplotlib inline
import scipy.stats as stats
from IPython.core.pylabtools import figsize
import numpy as np
figsize(16,8)

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

jet = plt.cm.jet
fig = plt.figure()
x = y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(x, y)

plt.subplot(121)
norm_x = stats.norm.pdf(x, loc=0, scale=1)
norm_y = stats.norm.pdf(y, loc=0, scale=1)
M = np.dot(norm_x[:, None], norm_y[None, :])
im = plt.imshow(M, interpolation='none', origin='lower',
                cmap=jet)

plt.xlim(0, 40)
plt.ylim(0, 40)
plt.title("Parameter Search Space for Normal Priors.")

ax = fig.add_subplot(122, projection='3d')
ax.plot_surface(X, Y, M, cmap=plt.cm.jet)
ax.view_init(azim=390)
plt.title("Parameter Search Space for Normal Priors");
```
